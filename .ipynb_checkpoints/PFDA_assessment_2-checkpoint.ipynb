## Looking at classifiers
***

The classifiers I will use for this project are Logistical regression, Random Forest, Decision Tree, Gaussian and K Nearest neighbour.

Logistic regression is a statistical method used for classification problems. It is a type of generalized linear model (GLM) that is used to model a binary outcome (i.e. two possible outcomes such as success/failure or true/false) as a function of one or more predictor variables. The logistic regression model predicts the probability that an instance belongs to a particular class (for example, the probability that a customer will buy a product given certain features of that customer). Given this predicted probability, a threshold is chosen, such as 0.5, to convert the predicted probabilities into class predictions.

Random forest is a supervised learning algorithm for classification and regression problems. It is an ensemble method, which means that it combines the predictions of multiple individual decision trees to create a more accurate and stable prediction. In a random forest, each decision tree is grown on a random sample of the data, and the features used to split each node in the tree are chosen at random from a subset of all features. This process helps to reduce overfitting and increase the overall performance of the model

A decision tree is a type of supervised machine learning used to categorize or make predictions based on how a previous set of questions were answered. The model is a form of supervised learning, meaning that the model is trained and tested on a set of data that contains the desired categorization.

The Gaussian Processes Classifier is a classification machine learning algorithm. Gaussian Processes are a generalization of the Gaussian probability distribution and can be used as the basis for sophisticated non-parametric machine learning algorithms for classification and regression

The basic concept of the k-NN algorithm is that an object is classified by a majority vote of its neighbors, with the object being assigned to the class most common among its k nearest neighbors. The k-NN algorithm is a "lazy learning" algorithm, which means that it does not build a model ahead of time, but rather waits to make predictions until a new input is encountered. To classify a new input, the algorithm simply looks at the k closest training examples to that input and assigns the new input the class label that is most common among those k closest examples.
